```python
import numpy as np
import matplotlib.pyplot as plt
from src.config import *

n_episodes = 9207070
epsilon_start = 1.0
epsilon_min = 0.005
decay_rate = 0.000005
rewards = []
actions_taken = []
epsilon_values = np.zeros(n_episodes)


# EPSILON_START = 1.0
# EPSILON_MIN = 0
# EPSILON_DECAY_RATE = 0.00001

epsilon_values = []


for episode in range(n_episodes):
    epsilon = epsilon_min + (epsilon_start - epsilon_min) * np.exp(-decay_rate * episode)
    epsilon_values.append(epsilon)

plt.plot(epsilon_values)
plt.xlabel("Episode")
plt.ylabel("Epsilon")
plt.title("Epsilon Exponential Decay Curve")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants for epsilon decay
EPSILON_START = 1.0
EPSILON_MIN = 0.005
EPSILON_DECAY_RATE = 0.00001
timesteps = 639895

# Calculate epsilon values over time
epsilon_values = EPSILON_MIN + (EPSILON_START - EPSILON_MIN) * np.exp(-EPSILON_DECAY_RATE * np.arange(timesteps))

# Plotting
plt.figure(figsize=(10,6))
plt.plot(epsilon_values, label="Epsilon Decay", color='blue')
plt.title('Epsilon Decay Over Time')
plt.xlabel('Timesteps')
plt.ylabel('Epsilon Value')
plt.grid(True)
plt.legend()
plt.show()
```
