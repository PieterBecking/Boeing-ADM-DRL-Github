```python
# %%
import pandas as pd
import matplotlib.pyplot as plt


# latest_folder = "scenario_folder_0858.json"

file_path = os.path.join(scenario_folder_path, latest_folder)

# extract the id from the file path
id = file_path.split('_')[-1].split('.')[0]
print(f"Scenario ID: {id}")



# Load the JSON data from the file
with open(file_path, 'r') as file:
    data = json.load(file)

# Extract the scenarios from the JSON data
scenarios = data['outputs']

print(f"Data Folder: {data_folder}")

def time_to_minutes(t):
    h, m = map(int, t.split(':'))
    return h * 60 + m

def calculate_scaled_slack_for_scenario(scenario_data):
    """
    Calculate a slack metric for the given scenario, scaled between 0 and 1.
    
    Key idea:
    - Determine the total scenario horizon from the earliest departure to the latest arrival across all flights.
    - For each aircraft:
      - If no flights: Slack = 1.0 (fully free aircraft)
      - If one flight: Slack = (Horizon - FlightDuration) / Horizon
      - If multiple flights: Slack = (Sum of gaps between consecutive flights) / Horizon
    
    The scenario slack is the average slack across all aircraft.

    This makes intuitive sense:
    - An aircraft with no flights contributes maximum slack (1.0).
    - An aircraft with tightly packed flights and no gaps will have slack close to 0.
    - Slack is now robustly scaled between 0 and 1.
    """
    flights = scenario_data['flights']
    aircraft_ids = scenario_data['aircraft_ids']

    # If there are no flights at all, scenario is fully free
    if len(flights) == 0:
        return 1.0

    # Identify the overall scenario horizon (earliest dep to latest arr)
    dep_times = []
    arr_times = []
    for f_id, f_data in flights.items():
        dep_times.append(time_to_minutes(f_data['DepTime']))
        arr_times.append(time_to_minutes(f_data['ArrTime']))
    earliest_dep = min(dep_times)
    latest_arr = max(arr_times)
    horizon = latest_arr - earliest_dep

    # Organize flights by aircraft
    aircraft_flight_map = {ac: [] for ac in aircraft_ids}
    for f_id, f_data in flights.items():
        dep = time_to_minutes(f_data['DepTime'])
        arr = time_to_minutes(f_data['ArrTime'])
        aircraft_flight_map[f_data['Aircraft']].append((f_id, dep, arr))

    # Sort each aircraft's flights by departure time
    for ac in aircraft_flight_map:
        aircraft_flight_map[ac].sort(key=lambda x: x[1])

    aircraft_slacks = []
    for ac, f_list in aircraft_flight_map.items():
        if len(f_list) == 0:
            # No flights -> fully free aircraft
            ac_slack = 1.0
        elif len(f_list) == 1:
            # One flight: Slack = (Horizon - FlightDuration) / Horizon
            flight_duration = f_list[0][2] - f_list[0][1]
            ac_slack = (horizon - flight_duration) / horizon if horizon > 0 else 0
        else:
            # Multiple flights: Slack = sum_of_gaps / Horizon
            gaps = []
            for i in range(len(f_list)-1):
                current_arr = f_list[i][2]
                next_dep = f_list[i+1][1]
                gap = next_dep - current_arr
                gaps.append(gap)
            total_gap = sum(gaps) if gaps else 0
            ac_slack = total_gap / horizon if horizon > 0 else 0

        aircraft_slacks.append(ac_slack)

    scenario_slack = sum(aircraft_slacks)/len(aircraft_slacks) if aircraft_slacks else 1.0
    return scenario_slack

#  


# Create a DataFrame to store scaled slack results for each scenario
scaled_slack_results = []
for scenario_name, scenario_data in scenarios.items():
    scenario_slack = calculate_scaled_slack_for_scenario(scenario_data)
    scaled_slack_results.append({"Scenario": scenario_name, "ScaledSlackMetric": scenario_slack})

scaled_slack_df = pd.DataFrame(scaled_slack_results)
print(f"The average slack metric across all scenarios is: {scaled_slack_df['ScaledSlackMetric'].mean():.2f}")
print(f"    The scenario with the highest slack metric is: {scaled_slack_df['Scenario'][scaled_slack_df['ScaledSlackMetric'].idxmax()]} with a slack metric of {scaled_slack_df['ScaledSlackMetric'][scaled_slack_df['ScaledSlackMetric'].idxmax()]:.2f}")
print(f"    The scenario with the lowest slack metric is: {scaled_slack_df['Scenario'][scaled_slack_df['ScaledSlackMetric'].idxmin()]} with a slack metric of {scaled_slack_df['ScaledSlackMetric'][scaled_slack_df['ScaledSlackMetric'].idxmin()]:.2f}")

# Plot the distribution of scaled slack across scenarios
plt.figure(figsize=(8,6))
plt.hist(scaled_slack_df['ScaledSlackMetric'], bins=10, edgecolor='black')
plt.title("Distribution of Scaled Slack Metrics Across Scenarios")
plt.xlabel("Scaled Slack Metric (0 to 1)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# %%
# Save the scaled slack metrics to a CSV file in the scenario folder path:
output_file_scaled = os.path.join(scenario_folder_path, f"scenario_scaled_slack_metrics_{id}.csv")
scaled_slack_df.to_csv(output_file_scaled, index=False)
print(f"Scaled slack metrics saved to {output_file_scaled}")
```
